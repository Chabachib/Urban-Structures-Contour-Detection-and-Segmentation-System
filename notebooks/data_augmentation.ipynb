```python
import os
import numpy as np
from PIL import Image
```

## Data Augmentation

```python
def flip_images(input_folder, output_folder):
    for file_name in os.listdir(input_folder):
        if file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
            input_path = os.path.join(input_folder, file_name)
            base_name, ext = os.path.splitext(file_name)
            new_file_name = f"{base_name}_flipped{ext}"
            output_path = os.path.join(output_folder, new_file_name)
            with Image.open(input_path) as img:
                flipped_img = img.transpose(Image.FLIP_LEFT_RIGHT)
                flipped_img.save(output_path)
            print(f"Processed: {file_name} -> {new_file_name}")
```

```python
input_real_folder = "../data/real" # Input of the real image data
output_real_folder = "../data/real" # Output for the augmented real image data
```

```python
input_gt_folder = "../data/ground-truth" # Input of the ground truth image data 
output_gt_folder = "../data/ground-truth" # Output for the augmented ground truth image data
```

```python
flip_images(input_real_folder, output_real_folder)
flip_images(input_gt_folder, output_gt_folder)
```
